<h2>Question 4</h2>

```python
x = np.arange(-16.001,16,0.001)
x_inv = np.arange(0.001,16,0.001)
```

```python
def exp_2(x):
    return 2**(x)
```

```python
def f_log(x):
    return math.log2(x)
```

```python
fig, f_plot = plt.subplots(1,1,figsize=(15,10), sharex=True)
# fig.text(0.05,0.5, "y", ha="center", va="center", rotation=90)
plt.xlabel("x - axis")
plt.ylabel("y - axis")

f_plot.grid(True)
# f_plot.title.set_text(r"$\bf{s_3$")
f_plot.set_xlim(-4,16)
f_plot.set_ylim(-4,16)

f = [ exp_2(i) for i in x ]
f_inverse = [ f_log(i) for i in x_inv ]
f_xy = x

# plots[:].set_xlim(5)
f_plot.plot( x, x*0, "k")
f_plot.plot( x*0, x, "k" )
f_plot.plot( x, f, "g", label="$f\ (x)=2^x$" )
f_plot.plot( x_inv, f_inverse, "r", label="$f^{\ -1\ }(x)=log_2(x)$" )
f_plot.plot( x, f_xy, label="$x\ =\ y$" )

#points to draw
#f
x_list = [0,1,2,3]
y_list = [ exp_2(x) for x in x_list ]
# print(y_list)

#f inverse
x_inv_list = [1,2,4,8]
y_inv_list = [ int(f_log(x)) for x in x_inv_list ]
# print(y_inv_list)

f_plot.scatter(x_list, y_list, marker="x", label="$f\ \ points$")
f_plot.scatter(x_inv_list, y_inv_list, marker="s", label="$f^{\ -1}\ points$")

for i, txt in enumerate(x_list):
    f_plot.annotate((x_list[i],y_list[i]), (x_list[i], y_list[i]))
    f_plot.annotate((x_inv_list[i],y_inv_list[i]), (x_inv_list[i], y_inv_list[i]))

f_plot.legend()
plt.xticks([i for i in range(-4,16,1)])

plt.title(r"$\bf{Plot\ of\ f(x)\ =\ 2^x,\ f^{-1}(x)\ =\ log_2(x),\ y\ =\ x\ and\ the\ respective\ points }$")
```
